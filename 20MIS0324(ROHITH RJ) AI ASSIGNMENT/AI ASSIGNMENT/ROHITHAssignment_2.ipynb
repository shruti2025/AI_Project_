THE OUTLIERS ARE REPLACED WITH MEDIAN METHOD

```python
q1 = df.Na_to_K.quantile(.25)
q3 = df.Na_to_K.quantile(.75)
```

```python
IQR = q3 - q1
IQR
```

```python
upper_limit = q3+1.5*IQR
lower_limit = q1-1.5*IQR
```

```python
df.median()
```

```python
df['Na_to_K'] =np.where(df['Na_to_K']>upper_limit,30.4,df['Na_to_K'])
```

***TASK 2:
DEVELOP THE MODEL***

```python
#BUILD AN ANN MODEL
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
     
```

```python
# Initializing the seq model
reg_model = Sequential()
# Adding the input layer to the model
reg_model.add(Dense(4,activation='relu'))
# Adding the 1st hidden layer to the model
reg_model.add(Dense(64,activation='relu'))
# Adding the 2nd hidden layer to the model
reg_model.add(Dense(32,activation='relu'))
# Adding output layer
reg_model.add(Dense(1,activation='softmax'))
```

```python
#compile the model
reg_model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
#train and test the model with the help of(xtrain,xtest,ytrain,ytest)
reg_model.fit(xtrain,ytrain,batch_size=4,epochs=15,validation_data=(xtest,ytest))
```

***TASK 3:
Test the model with random data***

```python
#Test with random data
out=reg_model.predict([[600,390,124,8,723]])
out
```
